```python
import json
import re
import random
```

```python
def load_rules(filepath="intent.json"):
    #load JSON file
    with open(filepath, "r") as file:
        data = json.load(file)
    return data["intents"]
```

```python
def generate_response(user_input, intents, memory):
    #checks intents and patterns for match 
    for intent in intents:
        for pattern in intent["patterns"]:

            match = re.search(pattern, user_input, re.IGNORECASE)

            if match:
                response = random.choice(intent["responses"])#random responses 
                groups = match.groupdict() #extract regex vaariable

               

                
                if "username" in groups:
                    memory["name"] = groups["username"]
                    return response.format(username=memory["name"])

               
                if "emotion1" in groups and "emotion2" in groups:
                    memory["emotions"] = [groups["emotion1"], groups["emotion2"]]
                    return response.format(
                        emotion1=memory["emotions"][0],
                        emotion2=memory["emotions"][1]
                    )

                #stores single emotions 
                if "emotion" in groups:
                    memory["emotion"] = groups["emotion"]

                #stores reason
                if "reason" in groups:
                    memory["reason"] = groups["reason"]

                #stores strategy
                if "strategy" in groups:
                    memory["strategies"].append(groups["strategy"])

                #stores feeling
                if "feeling" in groups:
                    memory["feeling"] = groups["feeling"]

                
                if "thought" in groups:
                    memory["thought"] = groups["thought"]

                

                if "thing" in groups:
                    memory["gratitude"] = groups["thing"]

                #sentiment classification
                if intent["tag"] == "sentiment_positive":
                    memory["sentiment"] = "positive"

                if intent["tag"] == "sentiment_negative":
                    memory["sentiment"] = "negative"

               
                if "goal" in groups:
                    memory["goal"] = groups["goal"]

               
                if "person" in groups:
                    memory["person"] = groups["person"]

                
                if "activity" in groups:
                    memory["activity"] = groups["activity"]

                #try inserting matched groups into responses 
                if groups:
                    try:
                        return response.format(**groups)
                    except:
                        pass  

               #farwell uses stored name if available
                if intent["tag"] == "farewell":
                    if memory["name"] is not None:
                        return response.format(username=memory["name"])
                    else:
                        return "Goodbye! Take care of yourself."

                return response
    #defualt response
    return "Tell me more about that."
```

```python
def chatbot():
    intents = load_rules() #load all patterns/responses 
    memory = {
        "name": None,
        "emotion": None,
        "emotions": [],
        "reason": None,
        "strategies": [],
        "feeling": None,
        "thought": None,
        "gratitude": None,
        "sentiment": None,
        "goal": None,
        "activity": None,
        "person": None
    }

    print("Eliza: Hello! I'm ready to chat. Type 'exit' to quit.\n")

    while True:
        user_input = input("You: ").strip()

        reply = generate_response(user_input, intents, memory)
        print("Eliza:", reply)

    #exit condition
        if re.search(r"\b(exit|quit|bye|goodbye)\b", user_input, re.IGNORECASE):
            break


chatbot()
```
